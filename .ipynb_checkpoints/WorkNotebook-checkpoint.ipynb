```python
import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
WITH FacCount AS (
SELECT memid,facid,COUNT(facid) AS MemFacCount FROM Bookings WHERE memid != 0 GROUP BY memid,facid
),
FacSum AS (
SELECT facid,SUM(MemFacCount) AS MemCountSum FROM FacCount GROUP BY facid
)

SELECT f.name,m.firstname,m.surname, ROUND(fc.MemFacCount * 100.0 / fs.MemCountSum,2) || '%' AS MemUsage
FROM FacCount fc
INNER JOIN Members m 
	ON m.memid = fc.memid
INNER JOIN Facilities f
    ON fc.facid = f.facid
INNER JOIN FacSum fs
    ON fc.facid = fs.facid
ORDER BY f.name
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def query(conn,query) :
    cur = conn.cursor()
    cur.execute(query)
    for row in cur.fetchall() :
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    print("2. Query all tasks")
    select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
